```python
import matplotlib.pyplot as plt #그래프
import seaborn as sns #데이터분석연습용 + 그래프
```

```python
# matplotlib #커스터마이징 가능,디자인적으로 어려움
# seaborn #자동으로 예쁜 그림 그려줌, 상상하는걸로 구현하기 어려움
# pandas #간단하게 그려볼 때

anscombe = sns.load_dataset("anscombe")
anscombe

#dataset 열이 1인 행만 추출해보세요.
anscombe[anscombe['dataset']=='I']

#dataset 열을 분류해보기
data1=anscombe[anscombe['dataset']=='I']
data2=anscombe[anscombe['dataset']=='II']
data3=anscombe[anscombe['dataset']=='III']
data4=anscombe[anscombe['dataset']=='IV']

#data 평균 구하기
print(data1.mean())
print(data2.mean())
print(data3.mean())
print(data4.mean())

#data 표준편차 구하기
print(data1.std())
print(data2.std())
print(data3.std())
print(data4.std())
```
