```python
total = []

for n in range(1, 10):                       #1회부터 순서대로 로또 정보 요청
  url=requests.get(f"https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q={n}%ED%9A%8C%EC%B0%A8+%EB%A1%9C%EB%98%90")
  html=BeautifulSoup(url.text)
  number=html.find("div",class_='lottonum').find_all("span")

  del number[-2]
  del number[-2]

  box = []
  for i in number: box.append(int(i.text))

  total.append(box)
  print(f"{n}회 로또 데이터 저장완료...{box} ")
  time.sleep(1)
```

```python
total = []

for n in range(1, 10):                       
  url=requests.get(f"https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C+{n}%ED%9A%8C")
  html=BeautifulSoup(url.text)
  number=html.find("td",class_='number').find_all("span")

  box = []
  for i in number: box.append(int(i.text))

  total.append(box)
  print(f"{n}회 연금복권 데이터 저장완료...{box} ")
  time.sleep(1)
```

```python
df2=pd.DataFrame(total, columns=['1번째','2번째','3번째','4번째','5번째','6번째'])
df2
df2.to_excel("pension.xlsx")
```
